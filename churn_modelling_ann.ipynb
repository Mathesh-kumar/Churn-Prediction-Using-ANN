### **Deep Learning model to predict the propensity of churn for each customer**.

### **Steps:**
1. Import Needed Libraries
2. Data Preprocessing
  1. Importing needed dataset
  2. Splitting dependent & independent variable
  3. Missing value treatment
  4. Encoding categorical data
  5. Splitting training and test set
  6. Feature scaling
3. Building ANN Model
  1. Initializing ann
  2. Adding input layer & first hidden layer
  3. Adding second hidden layer
  4. Adding output layer
4. Training ANN Model
  1. Compiling ann
  2. Training ann
5. Testing ANN Model
  1. Predicting test results
  2. Confusion matrix - accuracy

```
# Here we are going to build an artificial neural network model for churn modelling. 
# whether a customer is going to exit from the bank or not (churn prediction).
```

```
# 1. importing needed libraries

import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.preprocessing import LabelEncoder

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.metrics import confusion_matrix, accuracy_score
```

```
# 2. data preprocessing
# 2.1. importing needed dataset

df = pd.read_csv('Churn_Modelling.csv')
df.head()
```

```
# 2.2. splitting dependent and independent variables

# since we doesn't need RowNumber, CustomerId, Surname we dropped it.
x = df.iloc[:, 3:-1].values
x
```

```
y = df.iloc[:, -1].values
y
```

```
# 2.3. checking for missing values

df.isnull().sum().values
```

```
# 2.4. encoding categorical data
# in our independent variable x Geography & Gender are categorical.

# since gender is binary we are using label encoder
le = LabelEncoder()
x[:, 2] = le.fit_transform(x[:, 2])
x
```

```
# since geography is not binary we are using one hot encoding
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
x
```

```
# 2.5. splitting dataset into training and testing

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```
print(x_train)
print()
print(x_test)
```

```
# 2.6. feature scaling
# it is very important in deep learning. we have to scale all values to same range.

# to avoid information leakage x_train is fitted and transformed
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```
print(x_train)
print()
print(x_test)
```

```
# 3. building ann model
# 3.1. initializing ann

ann = tf.keras.models.Sequential()
```

```
# 3.2. adding input layer and first hidden layer

# there is no rule-of-thumb to fix units. you have add of your own.
# for fully connected neural network we have to use rectifier as activation function
# for non binary classification activation should be 'softmax'
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```
# 3.3. adding second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```
# 3.4. adding output layer

# since we have one output value to predict; units=1
# for binary output we can use either sigmoid/threshold as activation function
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```
# 4. training ann model
# 4.1. compiling ann

# adam optimizer - stochastic gradient descent
# loss - binary_crossentropy for binary output, categorical_crossentropy for non binary output
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
# 4.2. training ann

# batch learning is more efficient in ann
# high no of epochs we can get higher accuracy
ann.fit(x_train, y_train, batch_size=32, epochs=100)
```

```
# 5. making prediction and evaluating model

# 5.1. predicting test results
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5) # to convert probability value of output to binary
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```
# 5.2. cofusion matrix to see model accuracy

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```
# training accuracy - 0.8634; test accuracy - 0.86
# we got similar results from the test and train set, our model is predicting results effectively.
# the end of artificial neural network.
```
